```python
"""
multi_sector_johansen_pairstrat.py

Multisector cointegration strategy:
- Uses Johansen test on (sector ETFs + market ETF)
- Normalizes the chosen cointegration vector so market weight = -1
- Builds spread = w_sectors' * prices_sectors + w_market * price_market  (w_market = -1)
- Standardizes spread (rolling mean/std) -> zscore
- Trading rules: entry / exit on zscore
- Simple backtest (returns, transaction costs, sharpe, max drawdown)
Dependencies: pip install yfinance pandas numpy matplotlib statsmodels scipy
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from datetime import datetime

# --------------------------
# USER PARAMETERS
# --------------------------
market = "SPY"   # ETF mercato di riferimento
sector_etfs = [  # SPDR sector ETFs (es.)
    "XLB",  # Materials
    "XLE",  # Energy
    "XLF",  # Financials
    "XLI",  # Industrials
    "XLK",  # Technology
    "XLP",  # Consumer Staples
    "XLU",  # Utilities
    "XLV",  # Health Care
    "XLY",  # Consumer Discretionary
    "XLC"   # Communication Services
]
tickers = sector_etfs + [market]
start_date = "2018-01-01"
end_date = "2025-09-01"

# Johansen / VECM parameters
johansen_det_order = 0  # deterministic term: 0 = no deterministic term, 1 = constant
johansen_k_ar_diff = 1  # lag order in differences (small values ok; tune)

# trading params
roll_mean = 110   # days for zscore mean (approx quarterly trading)
roll_std = 110
entry_z = 1.5
exit_z = 0.4
tc = 0.0005      # transaction cost as fraction of notional traded
initial_capital = 1_000_000.0  # notional base for sizing
vol_target = 0.20  # target annual vol per strategy (for simple risk scaling)
annual_trading_days = 252

# --------------------------
# DATA FETCH
# --------------------------
print("Scarico dati da Yahoo Finance...")
data = yf.download(tickers, start=start_date, end=end_date, progress=False)['Close']
data = data.dropna(how='all').ffill().dropna()  # riempi e rimuovi righe vuote
prices = data[tickers].copy()
prices = prices.dropna()  # sincronizza
print(f"Dati scaricati: {prices.shape[0]} giorni su {len(tickers)} tickers")

# optionally, use log-prices for Johansen (often better behaved)
logp = np.log(prices)

# --------------------------
# JOHANSEN TEST
# --------------------------
# The Johansen implementation expects a numpy array (T x N)
# and returns eigenvectors (cointegration vectors) as columns in .eigvec
# We'll pick the first cointegration vector (most significant) by default.
Y = logp.values  # T x N

# run Johansen
print("Eseguo Johansen...")
jres = coint_johansen(Y, det_order=johansen_det_order, k_ar_diff=johansen_k_ar_diff)
# jres.evec is array N x N where columns are eigenvectors
eigvecs = jres.evec  # shape (N, N)
# pick first vector (index 0) which corresponds to the largest eigenvalue (strongest CI)
cvec = eigvecs[:, 0]  # length N

# map vector components to tickers
vec = pd.Series(cvec, index=tickers)
print("Vettore di cointegrazione (non normalizzato):")
print(vec.round(4))

# --------------------------
# NORMALIZZAZIONE: fissiamo peso su market = -1
# Spread = sum_i w_i * logP_i  ; set w_market = -1 so spread ~ sectors - beta * market
# If cvec_market is v_mkt, we compute w = -cvec / v_mkt  such that w_market = -1
# --------------------------
m_idx = tickers.index(market)
v_m = cvec[m_idx]
if abs(v_m) < 1e-8:
    raise RuntimeError("Il coefficiente di market nel vettore di cointegrazione è ~0; scegli un altro vettore o controlla i dati.")

w = -cvec / v_m
w_series = pd.Series(w, index=tickers)
print("\nVettore normalizzato (w) con peso mercato = -1:")
print(w_series.round(4))

# compute spread (use log-prices)
spread_log = (logp * w_series).sum(axis=1)  # s_t = w' * logP_t
# optionally convert to level spread: exp(spread_log) or keep log-spread for zscore (both fine)
spread = spread_log  # we'll work with log-spread

# --------------------------
# Z-SCORE (rolling)
# --------------------------
s_mean = spread.rolling(window=roll_mean, min_periods=20).mean()
s_std  = spread.rolling(window=roll_std, min_periods=20).std()
zscore = (spread - s_mean) / s_std

# --------------------------
# SIGNALS & POSITION SIZING
# - For simplicity we create a *market-neutral* relative position:
#   if z > entry -> SHORT spread (i.e. sell sectors basket, buy market)
#   if z < -entry -> LONG spread (buy sectors basket, sell market)
# - Convert weights to notional exposures (scale to target vol)
# --------------------------
signals = pd.Series(0, index=spread.index)
signals[zscore > entry_z] = -1
signals[zscore < -entry_z] = 1
# exit
exit_mask = (zscore.abs() < exit_z)
# We'll implement simple stateful position logic
positions_signal = pd.Series(0, index=signals.index)
pos = 0
for t in range(len(signals)):
    if pos == 0:
        if signals.iloc[t] != 0:
            pos = signals.iloc[t]
    else:
        if exit_mask.iloc[t]:
            pos = 0
    positions_signal.iloc[t] = pos

# Construct notional positions in each ticker from w_series:
# w gives weights for log-prices; but for exposure we need dollar exposures.
# We'll build a notional allocation vector N_t proportional to w_series (excluding market weight sign)
# Let basket B be sectors with weights w_sectors (positive means long sectors when spread long).
# For position = +1 (long spread) -> allocate notional such that strategy volatility ~ vol_target.
# We'll compute daily pnl from small-return approximation: pnl ≈ sum( w_dollar_i * ret_i )
# We'll scale w_dollar to target vol by historical realized vol of strategy returns unscaled.

# Build per-ticker returns
rets = prices.pct_change().fillna(0)

# define directionally consistent dollar weights per unit position:
# We want: for pos = +1 (long spread), dollar weight vector = +W (length N)
# where W = scale * w_dollars_raw and w_dollars_raw derived from w_series and current prices.
# Simpler approach: derive synthetic "strategy return per unit position" as:
# r_strategy_unit = sum( w_pct_i * ret_i ) where w_pct_i are percentage exposures summing to 0 (market-neutral)
# Build w_pct_i from sign(w_series) and magnitude |w| normalized.
w_abs = w_series.abs()
# keep signs: w_series positive means weight on ticker that contributes positively to spread
# compute raw percentages proportional to abs(w) but enforce net sum zero (market weight negative)
w_pct = w_series / w_abs.sum()  # this sums not to zero necessarily
# Better: compute raw dollar pattern proportional to w (so market negative)
# Normalize by sum of absolute to get bounded exposures
w_dollars_raw = w_series / w_series.abs().sum()

# compute synthetic series of returns for "unit notional" = sum(w_dollars_raw * returns)
r_unit = (rets * w_dollars_raw).sum(axis=1)

# estimate current realized vol of r_unit (annualized)
rolling_vol = r_unit.rolling(63).std() * np.sqrt(annual_trading_days)
# scale factor to reach vol_target (avoid division by 0)
scale = (vol_target / rolling_vol).replace([np.inf, -np.inf], np.nan).fillna(0.0)

# position notional at time t for pos = +1 is: scale_t * initial_capital * w_dollars_raw
# similarly for pos = -1 invert sign
# We'll compute daily PnL using this rule.
notional = pd.DataFrame(0.0, index=prices.index, columns=tickers)
for t, date in enumerate(prices.index):
    s = positions_signal.iloc[t]
    if s == 0:
        continue
    sc = scale.iloc[t] if not np.isnan(scale.iloc[t]) else 0.0
    base = sc * initial_capital
    notional.loc[date] = base * (s * w_dollars_raw)  # sign included

# PnL: notional shifted * returns (we assume rebalanced daily at close)
pnl = (notional.shift(1) * rets).sum(axis=1).fillna(0.0)  # PnL realized next day from yesterday's notional
# transaction costs: when notional changes, pay tc * sum(|delta notional|)
txn = tc * notional.diff().abs().sum(axis=1).fillna(0.0)
pnl_net = pnl - txn
cumret = (1 + pnl_net / initial_capital).cumprod() - 1

# performance metrics
def annualize_return(r_series):
    total_days = len(r_series)
    total_return = (1 + r_series).prod() - 1
    years = total_days / annual_trading_days
    return (1 + total_return) ** (1 / years) - 1 if years>0 else np.nan

daily_rets = pnl_net / initial_capital
ann_return = annualize_return(1 + daily_rets)
ann_vol = daily_rets.std() * np.sqrt(annual_trading_days)
sharpe = (daily_rets.mean() / daily_rets.std()) * np.sqrt(annual_trading_days) if daily_rets.std() != 0 else np.nan
max_dd = (cumret.cummax() - cumret).max()

print("\nPerformance summary:")
print(f"Period: {prices.index[0].date()} to {prices.index[-1].date()}")
print(f"Cumulative return (strategy): {cumret.iloc[-1]:.2%}")
print(f"Annualized return (approx): {ann_return:.2%}")
print(f"Annualized vol (approx): {ann_vol:.2%}")
print(f"Sharpe (ann): {sharpe:.2f}")
print(f"Max drawdown: {max_dd:.2%}")

# --------------------------
# PLOTS
# --------------------------
plt.figure(figsize=(12,4))
plt.plot(spread, label="Spread (log, cointegrated linear combo)")
plt.plot(s_mean, label=f"Rolling mean ({roll_mean})", alpha=0.8)
plt.title("Spread (cointegration vector)"); plt.legend(); plt.show()

plt.figure(figsize=(12,4))
plt.plot(zscore, label="Z-score")
plt.axhline(entry_z, ls='--', label='entry')
plt.axhline(-entry_z, ls='--')
plt.axhline(exit_z, ls=':', label='exit')
plt.axhline(-exit_z, ls=':')
plt.title("Z-score dello spread"); plt.legend(); plt.show()

plt.figure(figsize=(12,4))
plt.plot(cumret, label="Equity curve (cum ret)")
plt.title("Equity curve"); plt.legend(); plt.show()

# Optional: show weights derived from w_series
print("\nPeso normalizzato w (interpretabile come log-weight):")
print(w_series.round(4))

# Save outputs
out = pd.DataFrame({
    'spread': spread,
    'zscore': zscore,
    'signal': positions_signal,
    'pnl': pnl,
    'pnl_net': pnl_net,
    'cumret': cumret
})
out.to_csv("multisector_johansen_results.csv")
print("\nRisultati salvati in multisector_johansen_results.csv")
```
